```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes  import GaussianNB


df=pd.read_csv("social_media_screentime_mental_health_2026.csv")
df=df.drop('participant_id',axis=1)
x=df.drop(['wellbeing_band',axis=1)
y=df['wellbeing_band']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
numeric_features=[col for col in x.columns if x[col].dtypes!='O']
numeric_features
categorical_features=[col for col in x.columns if x[col].dtypes=='O']
categorical_features

numeric_pipeline=Pipeline([("imputer",SimpleImputer(strategy='mean')),("scaler",StandardScaler())])
categorical_pipiline=Pipeline([("imputer",SimpleImputer(strategy='most_frequent')),("encoder",OneHotEncoder(handle_unknown='ignore'))])
preprocessor=ColumnTransformer([("numrical",numeric_pipeline,numeric_features),('categorical',categorical_pipiline,categorical_features)])
x_train_preprocessed=preprocessor.fit_transform(x_train)
x_test_preprocessed=preprocessor.transform(x_test)
logis=LogisticRegression(max_iter=1000)
svc=SVC(probability=True)
gnb=GaussianNB()
logis.fit(x_train_preprocessed,y_train)
svc.fit(x_train_preprocessed,y_train)
gnb.fit(x_train_preprocessed,y_train) 
models={
    'Logistic_Regression':logis,
    'SVC':svc,
    'GaussianNB':gnb
}
for name,model in models.items():
    y_pred=model.predict(x_test_preprocessed)

    acc=accuracy_score(y_test,y_pred)

    print(f"{name}:{acc}")

from sklearn.model_selection import cross_val_score
for name, model in models.items():
    scores = cross_val_score(
        model,
        x_train_preprocessed,
        y_train,
        cv=5,
        scoring="accuracy"
    )
    print(
        name,
        "Mean CV Accuracy:",
        scores.mean()
    )
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
from sklearn.model_selection import GridSearchCV
param_grid={
     'kernel':['rbf','linear','poly'],
     'C':[0.01,0.1,1,10,100],
     'gamma':['scale','auto'],
     'degree':[2,3,4]
}
grid=GridSearchCV(SVC(probability=True),param_grid=param_grid,cv=cv,scoring='accuracy',n_jobs=-1)
grid.fit(x_train_preprocessed,y_train)
grid.best_params_
model=grid.best_estimator_
 
y_pred=model.predict(x_test_preprocessed)
acc=accuracy_score(y_test,y_pred)
print(f"accuracy:{acc}")
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
from sklearn.model_selection import cross_val_score

scores = cross_val_score(
    model,
    x_train_preprocessed,
    y_train,
    cv=cv,
    scoring="accuracy"
)

print("Individual CV Scores:", scores)

print(
    f"Mean CV Score: "
    f"{scores.mean() * 100:.2f}%"
)

print(
    f"Standard Deviation: "
    f"{scores.std() * 100:.2f}%"
)
import pickle
pickle.dump(model,open("model.pkl",'wb'))
```
